```python
df_diabetes = new_df[new_df['diabetes'] == 1]
```

```python
df_diabetes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=df_diabetes, x='age', bins=15, kde=True)#I'm telling the sns that it should only use age
plt.title('Age Distribution of Diabetic Patients')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
#let's see tthe diference in the a diabetic patiente and a non diabetic patiente regarding stab.glu and glyhb
```

```python
sns.boxplot(x='diabetes', y='stab.glu', data=df)
plt.title('Glucose Levels by Diabetes Status')
plt.show()
```
